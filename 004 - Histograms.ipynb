Histogramas son buenos para visualizar componentes individuales de color

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt #librería de visualización de graficos

image=cv2.imread("images/park/4468367264_87a9d5049f_b.jpg")
histogram=cv2.calcHist([image],[0],None,[256],[2,256]) #solo para el color azul

#mostramos el histograma, ravel() convierte el array en 1 dimensión
plt.hist(image.ravel(),256,[0,256]);
plt.show()

#ver canales de color separados
color=('b','g','r')

#ahora se separan los colores y se grafica cada uno
for i, col in enumerate(color):
    histogram2=cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram2,color=col)
    plt.xlim([0,256])

plt.show()

cv2.imshow("image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

**cv2.calcHist(images,channels,mask,histSize,ranges[,hist[,accumulate]])**

* images: it is the source image of type uint8 or float32. always in []

* channels: Also in [], index of channels for which we calculate histogram. for grayscale is [0],for color  can pass [0],[1],[2].

* mask: If you want to find histogram of particular region, you have to create a mask image for that and give it a mask.

* histSize: this represent our BIN count. Always in [], for scale we pass [256]

* ranges: Normally [0,256]
